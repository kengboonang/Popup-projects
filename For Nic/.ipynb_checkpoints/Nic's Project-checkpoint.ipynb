```python
# loading workbook
import openpyxl as xl
import win32com.client
import os
import sys
import PIL
from PIL import ImageGrab, Image
```

```python
# creating one word doc for all rows
import datetime
from docxtpl import DocxTemplate, InlineImage
from docx.shared import Cm

workbook = xl.load_workbook('automobile_data.xlsx')
sheet_1 = workbook['automobile_data']
template = DocxTemplate('template.docx')
table_contents = []

# reading values from excel
for i in range(2, sheet_1.max_row + 1):
    table_contents.append({
        'Index': i - 1,
        'company': sheet_1.cell(i, 1).value,
        'Body': sheet_1.cell(i, 2).value,
        'Wheel': sheet_1.cell(i, 3).value,
        'length': sheet_1.cell(i, 4).value
    })

# settling image
image = InlineImage(template,'1.png',Cm(10))

# transferring to word doc
context = {
    'title': 'Automated Report',
    'day': datetime.datetime.now().strftime('%d'),
    'month': datetime.datetime.now().strftime('%b'),
    'year': datetime.datetime.now().strftime('%Y'),
    'table_contents': table_contents,
    'image': image
}

template.render(context)
template.save('Automated_report.docx')
```

```python
# creating new word doc for each row
import datetime
from docxtpl import DocxTemplate

template = DocxTemplate('template.docx')
table_contents = []

# reading values from excel
for i in range(2, sheet_1.max_row + 1):
    table_contents.append({
        'Index':   i - 1,
        'company': sheet_1.cell(i, 1).value,
        'Body':    sheet_1.cell(i, 2).value,
        'Wheel':   sheet_1.cell(i, 3).value,
        'length':  sheet_1.cell(i, 4).value
    })

# transferring to word doc
for i in range(len(table_contents)):
    file_name = str(table_contents[i]['Index']) + '.docx'
    context = {
        'title':          str(table_contents[i]['Index']),
        'day':            datetime.datetime.now().strftime('%d'),
        'month':          datetime.datetime.now().strftime('%b'),
        'year':           datetime.datetime.now().strftime('%Y'),
        'table_contents': [table_contents[i]]
    }
    
    template.render(context)
    template.save(file_name)
```

```python
# code for template

import win32com.client
import os
import sys
import PIL
from PIL import ImageGrab, Image

```

```python
# loading workbook
import openpyxl as xl
workbook = xl.load_workbook('data\Análise de Risco (Risk Analysis) Spreadsheet.xlsx')
sheet_1 = workbook['Sheet1']

# creating new word doc for each row
import datetime
from docxtpl import DocxTemplate, RichText

# loading template
template = DocxTemplate('templates\RA2.docx')
table_contents = []


# reading values from excel
for i in range(2, sheet_1.max_row + 1):
    table_contents.append({
        'ReportNumber':     sheet_1.cell(i, 1).value,
        'Location':         sheet_1.cell(i, 2).value,
        'Equipment':        sheet_1.cell(i, 3).value,
        'FabricationDate':  sheet_1.cell(i, 4).value,
        'EquipmentTag':     sheet_1.cell(i, 5).value,
        'Manufacturer':     sheet_1.cell(i, 6).value,
        'Model':            sheet_1.cell(i, 7).value,
        'SerialNumber':     sheet_1.cell(i, 8).value,
        'IndexNumber':      sheet_1.cell(i, 9).value,
        'MachineType':      sheet_1.cell(i, 10).value,
        'Rotation':         sheet_1.cell(i, 11).value,
        'Capacity':         sheet_1.cell(i, 12).value,
        'Power':            sheet_1.cell(i, 13).value,
        'Weight':           sheet_1.cell(i, 14).value,
        'Dimension':        sheet_1.cell(i, 15).value,
        'System':           sheet_1.cell(i, 16).value,
        'Other':            sheet_1.cell(i, 17).value,
        'Risks1':           sheet_1.cell(i, 18).value,
        'Location1':        sheet_1.cell(i, 19).value,
        'Table1':           sheet_1.cell(i, 20).value,
        'Table2':           sheet_1.cell(i, 21).value,
        'RiskLevel':        sheet_1.cell(i, 22).value,
        'Reccommendation':  sheet_1.cell(i, 23).value,
        'Risks2':           sheet_1.cell(i, 24).value,
        'Location2':        sheet_1.cell(i, 25).value,
        'Table1_1':         sheet_1.cell(i, 26).value,
        'Table2_1':         sheet_1.cell(i, 27).value,
        'RiskLevel_1':      sheet_1.cell(i, 28).value,
        'Reccommendation1': sheet_1.cell(i, 29).value,
        'Risks3':           sheet_1.cell(i, 30).value,
        'Location3':        sheet_1.cell(i, 31).value,
        'Table1_2':         sheet_1.cell(i, 32).value,
        'Table2_2':         sheet_1.cell(i, 33).value,
        'RiskLevel_2':      sheet_1.cell(i, 34).value,
        'Reccommendation2': sheet_1.cell(i, 35).value,
        'Risks4':           sheet_1.cell(i, 36).value,
        'Location4':        sheet_1.cell(i, 37).value,
        'Table1_3':         sheet_1.cell(i, 38).value,
        'Table2_3':         sheet_1.cell(i, 39).value,
        'RiskLevel_3':      sheet_1.cell(i, 40).value,
        'Recommendation3':  sheet_1.cell(i, 41).value,
        'DrawingNumber':    sheet_1.cell(i, 42).value,
        'Observations':     sheet_1.cell(i, 43).value,
        'Date':             sheet_1.cell(i, 44).value
    })

# transferring to word doc
for i in range(len(table_contents)):
    file_name = str(table_contents[i]['ReportNumber']) + '.docx'
    for var,value in table_contents[i].items():
        rt = RichText()
        # splitting by language
        if '/' in str(value):
            string = value.split('/')
            portuguese = string[0] + '/'
            english = string[1]
            rt.add(portuguese, color='#000000', size= 14)
            rt.add(english, color='#00B0F0', size= 14)
            table_contents[i][var] = rt
            
    context = {
        'table_contents': [table_contents[i]]
    }
    
    template.render(context)
    template.save(file_name)
```
